## Spliting data and handling inbalanced data useing Oversampling using SMOTE technique

```python
# dividing the X and the Y from the dataset
X = data.drop(['Class'], axis = 1)
Y = data["Class"]
```

```python
import pandas as pd
from imblearn.over_sampling import SMOTE

# Assuming 'X' is your feature matrix and 'y' is your target vector
# Instantiate SMOTE
smote = SMOTE(random_state=42)

# Fit and resample the whole dataset
X_resampled, y_resampled = smote.fit_resample(X, Y)

# Convert X_resampled and y_resampled back to a DataFrame if necessary
X_resampled_df = pd.DataFrame(X_resampled, columns=X.columns)
y_resampled_series = pd.Series(y_resampled, name='target')
```

## Concatenating X_resampled and y_resampled into single DataFrame to visualize correlation matrix

```python
# Concatenate X_resampled and y_resampled into a single DataFrame
new_data = pd.concat([X_resampled_df, y_resampled_series], axis=1)

# Calculate correlation matrix
correlation_matrix = new_data.corr()
fig = plt.figure(figsize = (10, 7))
sns.heatmap(correlation_matrix, vmax = .8, square = True )
plt.show()
```

```python
print('Distribution of the Classes in the subsample dataset')
print(new_data['target'].value_counts()/len(new_data))
colors = ["#1f77b4", "#ff7f0e"]  # Blue for class 0, Orange for class 1

# Plot countplot with specified colors
sns.countplot(x='target', data = new_data, palette=colors)
plt.title('Equally Distributed Classes', fontsize=14)
plt.show()
```

## Splitting data into X, Y and further train, test

```python
# dividing the X and the Y from the dataset
X = new_data.drop(['target'], axis = 1)
Y = new_data["target"]
```

```python
from sklearn.model_selection import train_test_split
xTrain,xTest,yTrain,yTest = train_test_split(X,Y,test_size=0.20,
                                                 random_state=42)
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
from sklearn.tree import DecisionTreeClassifier
dt_classifier = DecisionTreeClassifier()
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
```

```python

# Create a Voting Classifier with estimators as a list of tuples
voting_classifier = VotingClassifier(estimators=[
    ('Random Forest', rfc),
    ('Decision Tree', dt_classifier),
    ('Logistic Regression', log_reg),
], voting='hard')  # 'hard' voting means a majority vote
```

```python
# Fit the voting classifier
voting_classifier.fit(xTrain, yTrain)

# Make predictions using the voting classifier
yPred_Voting = voting_classifier.predict(xTest)
```

```python
# Now you can evaluate the voting classifier as well

from sklearn.metrics import classification_report, accuracy_score
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import f1_score, matthews_corrcoef
from sklearn.metrics import confusion_matrix

# n_outliers = len(fraud)
n_errors = (yPred_Voting != yTest).sum()
print("Model's Performance")

acc = accuracy_score(yTest, yPred_Voting)
print("The accuracy is {}".format(acc))

prec = precision_score(yTest, yPred_Voting)
print("The precision is {}".format(prec))

rec = recall_score(yTest, yPred_Voting)
print("The recall is {}".format(rec))

f1 = f1_score(yTest, yPred_Voting)
print("The F1-Score is {}".format(f1))

MCC = matthews_corrcoef(yTest, yPred_Voting)
print("The Matthews correlation coefficient is{}".format(MCC))

# printing the confusion matrix
LABELS = ['Normal', 'Fraud']
conf_matrix = confusion_matrix(yTest, yPred_Voting)
plt.figure(figsize=(8, 8))
sns.heatmap(conf_matrix, xticklabels=LABELS,
            yticklabels=LABELS, annot=True, fmt="d");
plt.title("Confusion matrix")
plt.ylabel('True class')
plt.xlabel('Predicted class')
plt.show()
```
